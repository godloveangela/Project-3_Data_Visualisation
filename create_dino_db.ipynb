```python
import sqlite3
import pandas as pd
import json
```

```python
# open JSON data
with open("data/dino_data.json") as f:
    data = json.load(f)

# # create a dataframe from the JSON data
# print(json.dumps(data, indent=4))
data_json = json.load(open('data/dino_data.json'))
```

```python
# create dataframes
df=pd.DataFrame(data_json['records'])
df_new=df[['eag','lag','phl','cll','odl','fml','gnl','lng','lat']]
df_final=df_new.rename(columns={'eag':'earliest-appearance','lag':'latest-appearance','phl':'phylum','cll':'class','odl':'order','fml':'family','gnl':'genus','lng':'longitude','lat':'latitude'})
df_final.head()
```

```python
# open JSON data
with open("data/taxonomy_data.json") as f:
    data = json.load(f)

# print(json.dumps(data,indent=4))

df = pd.DataFrame(data)
df1=df.rename(columns={0:'full_scientific_name', 1:'genus', 2:'species', 3:'group'})
df1.head()
```

```python
# merge the dataframes
df_merged = df_final.merge(df1, how='left', on='genus')
df_dropped=df_merged.drop(columns=['order', 'family', 'full_scientific_name', 'species'])
df_drop_dup= df_dropped.drop_duplicates()
df_drop_dup.head()
```
